# World Cloud for LinkedIn Cover Photo

## Import Packages

```python
import numpy as np
import os
import seaborn as sns
import wordcloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from PIL import Image
from wordcloud import WordCloud, ImageColorGenerator
```

## Set Folder Path

```python
path = os.getcwd()
```

## Import the Resume.txt file

```python
with open(path+'\\Resume.txt', 'r') as file:
    data = file.read().replace('\n', '')
```

## Tokenized and Remove the Stopwords

```python
stop_words = set(stopwords.words('english'))
newdata=word_tokenize(data)
filtered=[]
for w in newdata: 
    if w not in stop_words: 
        filtered.append(w)
```

## Build Word Cloud

```python
Res_AN=' '.join(filtered).upper()
G = np.array(Image.open(path+'\\Thinking-Bubble.png'))
sns.set(rc={'figure.figsize':(42,11)})
wordcloud = WordCloud(max_font_size=300, max_words=600,mask=G,background_color="white",contour_width=3, contour_color='black').generate(Res_AN)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.savefig(path+'\\Word-Cloud.png',bbox_inches='tight')
plt.show()
plt.close()
```
